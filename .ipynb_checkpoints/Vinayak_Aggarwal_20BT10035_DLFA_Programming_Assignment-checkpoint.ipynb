# PROBLEM 1.
Implementing a MP Neuron model that takes binary inputs and produces a binary output

```python
# assuming the input_types to be 1, -1 for excitatory and inhibitory
# output types to be 1, 0 for HIGH and LOW
class MPNeuron:
    def __init__(self, threshold, input_types):
        self.threshold = threshold
        self.input_types = input_types
    def compute_output(self, inputs):
        weighted_sum = 0
        for inp_type, inp in zip(self.input_types, inputs):
            weighted_sum += inp_type * inp
        return int(weighted_sum >= self.threshold )
```

###### For a single neuron having two inputs, only three possiblities exist.
1. Inhibitory - Inhibitory
2. Excitatory - Inhibitory
3. Excitatory - Excitatory

###### Testing it with 1

```python
n1 = MPNeuron(0, [-1, -1])
print("Threshold:", 0)
print("inputs: -1, -1 | expected: HIGH | output: ", n1.compute_output([-1, -1]))
print("inputs: -1, +1 | expected: HIGH | output: ", n1.compute_output([-1, 1]))
print("inputs: +1, -1 | expected: HIGH | output: ", n1.compute_output([1, -1]))
print("inputs: +1, +1 | expected: LOW  | output: ", n1.compute_output([1, 1]))

print()

n2 = MPNeuron(2, [-1, -1])
print("Threshold:", 2)
print("inputs: -1, -1 | expected: HIGH | output: ", n2.compute_output([-1, -1]))
print("inputs: -1, +1 | expected: LOW  | output: ", n2.compute_output([-1, 1]))
print("inputs: +1, -1 | expected: LOW  | output: ", n2.compute_output([1, -1]))
print("inputs: +1, +1 | expected: LOW  | output: ", n2.compute_output([1, 1]))

print() 

n3 = MPNeuron(-2, [-1, -1])
print("Threshold:", -2)
print("inputs: -1, -1 | expected: HIGH  | output: ", n3.compute_output([-1, -1]))
print("inputs: -1, +1 | expected: HIGH  | output: ", n3.compute_output([-1, 1]))
print("inputs: +1, -1 | expected: HIGH  | output: ", n3.compute_output([1, -1]))
print("inputs: +1, +1 | expected: HIGH  | output: ", n3.compute_output([1, 1]))
```

Similarly, can be shown for the other two cases as well

#### Part 1: AND logic gate
With simple reasoning, it can be deduced that an AND gate should be a single neuron with both the inputs of the type excitatory and threshold set to 2

```python
Neuron_AND = MPNeuron(2, [1, 1])
print("inputs: -1, -1 | expected: LOW  | output: ", Neuron_AND.compute_output([-1, -1]))
print("inputs: -1, +1 | expected: LOW  | output: ", Neuron_AND.compute_output([-1, 1]))
print("inputs: +1, -1 | expected: LOW  | output: ", Neuron_AND.compute_output([1, -1]))
print("inputs: +1, +1 | expected: HIGH | output: ", Neuron_AND.compute_output([1, 1]))
```

#### Part 2: OR logic gate
With simple reasoning, it can be deduced that an OR gate should be a single neuron with both the inputs of the type excitatory and threshold set to 0

```python
Neuron_OR = MPNeuron(0, [1, 1])
print("inputs: -1, -1 | expected: LOW  | output: ", Neuron_OR.compute_output([-1, -1]))
print("inputs: -1, +1 | expected: HIGH | output: ", Neuron_OR.compute_output([-1, 1]))
print("inputs: +1, -1 | expected: HIGH | output: ", Neuron_OR.compute_output([1, -1]))
print("inputs: +1, +1 | expected: HIGH | output: ", Neuron_OR.compute_output([1, 1]))
```

#### Part 3: NOT logic gate
1. NOT gate means for an input A, output should be ~A.
2. for an excitatory A, output should be 0
3. for an inhibitory A, output should be 1
4. Just one input is to be set, of the type inhibitory, threshold to be 0. this way, It will flip the input and output HIGH if the input was -1, else will output LOW

```python
Neuron_NOT = MPNeuron(0, [-1])
print("input: -1 | expected: HIGH | output: ", Neuron_NOT.compute_output([-1]))
print("input: +1 | expected: LOW  | output: ", Neuron_NOT.compute_output([1]))
```

#### Part 4: NOR logic gate

##### 1. We can either use just one neuron that behaves similar to AND gate

```python
Neuron_NOR = MPNeuron(2, [-1, -1])
print("inputs: -1, -1 | expected: HIGH | output: ", Neuron_NOR.compute_output([-1, -1]))
print("inputs: -1, +1 | expected: LOW  | output: ", Neuron_NOR.compute_output([-1, 1]))
print("inputs: +1, -1 | expected: LOW  | output: ", Neuron_NOR.compute_output([1, -1]))
print("inputs: +1, +1 | expected: LOW  | output: ", Neuron_NOR.compute_output([1, 1]))
```

##### 2. Or, we can pass the output of OR neuron to the NOT Neuron

```python
print("inputs: -1, -1 | expected: HIGH | output: ", Neuron_NOT.compute_output([Neuron_OR.compute_output([-1, -1])]))
print("inputs: -1, +1 | expected: LOW  | output: ", Neuron_NOT.compute_output([Neuron_OR.compute_output([-1, 1])]))
print("inputs: +1, -1 | expected: LOW  | output: ", Neuron_NOT.compute_output([Neuron_OR.compute_output([1, -1])]))
print("inputs: +1, +1 | expected: LOW  | output: ", Neuron_NOT.compute_output([Neuron_OR.compute_output([1, 1])]))
```

#### Part 5: 
Boolean function for a scenario when a staircase light is controlled by two switches such that light is ON only when one of the switches is ON and the othere is OFF

1. it is given that when both the lights are on, the bulb should remain off
2. it is given that when both the lights are off, the bulb should remain off
3. it is given that when one light is on other is off, the bulb should turn on
4. statement 1, 3 can be achieved by taking a NOT on AND neuron with both inputs treated excitatory
5. statement 2 in this setup will however result in the bulb to turn on which we do not want
6. we can pass the NOT + AND input to another AND gate this time coupled with an OR filtered output from the original inputs

```python
def query_part_5(input_1 = 1, input_2 = -1):
    output_AND1 = Neuron_AND.compute_output([input_1, input_2])
    output_NOT = Neuron_NOT.compute_output([output_AND1])
    output_OR = Neuron_OR.compute_output([input_1, input_2])
    return Neuron_AND.compute_output([output_NOT, output_OR])
```

```python
print("inputs: +1, +1 | expected: LOW  | output: ", query_part_5(1, 1))
print("inputs: +1, -1 | expected: HIGH | output: ", query_part_5(1, -1))
print("inputs: -1, +1 | expected: HIGH | output: ", query_part_5(-1, 1))
print("inputs: -1, -1 | expected: LOW  | output: ", query_part_5(-1, -1))
```

#### Part 6: 
On looking at the Truth Table of the function, it resembles the XOR logic gate.

A computationally efficient model will be the one that was also discussed in the class, that is a layer of 4 neurons and an output neuron having those 4 corresponding inputs

```python
# I1 | I2 | H1 | H2 | H3 | H4 | Out
# _________________________________
# +1 | +1 | +1 | 0  | 0  | 0  | 0
# _________________________________
# +1 | -1 | 0  | +1 | 0  | 0  | 1
# _________________________________
# -1 | +1 | 0  | 0  | +1 | 0  | 1
# _________________________________
# -1 | -1 | 0  | 0  | 0  | +1 | 0
```

Thus, we can individually deduce the H1, H2, H3, H4 gates first by looking at H as a function of I1, I2

```python
Neuron_NAND = MPNeuron(2, [-1, -1])
```

```python
def XOR(input_1 = 1, input_2 = -1):    
    H1 = Neuron_AND
    H2 = MPNeuron(2, [1, -1])
    H3 = MPNeuron(2, [-1, 1])
    H4 = Neuron_NAND
    hidden_layer = [H1, H2, H3, H4]
    Output = MPNeuron(1, [-1, 1, 1, -1])
    outputs_from_hidden = []
    for hidden_neuron in hidden_layer:
        output = hidden_neuron.compute_output([input_1, input_2])
        outputs_from_hidden.append(output)
    return Output.compute_output(outputs_from_hidden)
```

```python
print("inputs: +1, +1 | expected: LOW  | output: ", XOR(+1, +1))
print("inputs: +1, -1 | expected: HIGH | output: ", XOR(+1, -1))
print("inputs: -1, +1 | expected: HIGH | output: ", XOR(-1, +1))
print("inputs: -1, -1 | expected: LOW  | output: ", XOR(-1, -1))
```

# PROBLEM 2.
PCA from scratch, MNIST Dataset

```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
```

#### Part 1

```python
mnist = tf.keras.datasets.mnist
(train_X, train_Y), (test_X, test_Y) = mnist.load_data()
```

```python
image_1 = train_X[0]
image_1.shape
```

#### Part 2

```python
np.random.seed(42)
```

```python
random_indices = np.random.choice(len(train_X), 10000, replace=False)
chosen_images = train_X[random_indices]
flattened_images = chosen_images.reshape(-1, 28*28)
flattened_images.shape
```

Here, the flattened_images matrix is a 10000x784 shaped matrix with 10,000 images, one image per row, where a single image is represented by a 1D numpy array of 784 pixel values

#### Part 3

To standardize a 2D matrix means to normalize each and every column of the matrix, between 0 and 1 i.e. x_j = (x_j-mean_i)/(std_i)

```python
def standardize(x):    
    mean = np.mean(x, axis=0)
    std = np.std(x, axis=0)
    z = (x-mean)/np.maximum(std, 1e-18) # to avoid division by 0
    return z
```

#### Part 4

Steps to perform PCA:
1. Standardize the matrix
2. Compute the covariance matrix
3. Obtain the eigenvectors of the covariance matrix
4. Sort the eigenvalues to obtain the principal components explaining the maximum variances

```python
standardized_flattened_images = standardize(flattened_images)
covariance_matrix = np.cov(standardized_flattened_images.T)
eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix)
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
projection_matrix = eigenvectors[:, idx]
```

#### Part 5

```python
def reconstruct_data(P, Z, mean, std):
    X_reduced = np.dot(Z, P.T)
    X_reconstructed = (X_reduced * std) + mean
    return X_reconstructed
```

```python
X = flattened_images
P = projection_matrix
Z = standardized_flattened_images

mean = np.mean(flattened_images, axis=0)
std = np.std(flattened_images, axis=0)
errors = []

for p in range(24, 785, 20):
    P_reduced = P[:, :p]
    Z_reduced = np.dot(Z, P_reduced)
    X_reconstructed = reconstruct_data(P_reduced, Z_reduced, mean, std)
    
    first_image_original = X[0].reshape(28, 28)
    first_image_reconstructed = X_reconstructed[0].reshape(28, 28)
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(first_image_original, cmap='gray')
    plt.title('Original Image')
    plt.subplot(1, 2, 2)
    plt.imshow(first_image_reconstructed, cmap='gray')
    plt.title(f"Reconstructed Image with p={p}")
    plt.show()

    error = np.linalg.norm(X - X_reconstructed)
    errors.append(error)
    print(f"Reconstruction Error with p={p}: {error}")
```

#### Part 6

```python
components = list(range(24, 785, 20))

plt.figure(figsize=(10, 6))
plt.plot(components, errors, marker='o')
plt.title('Reconstruction Error vs Number of Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Reconstruction Error')
plt.grid(True)
plt.show()
```
